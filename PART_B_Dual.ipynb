```python
from myMatrixLpSolver import lp_optimize
from gurobipy import *
import numpy as np
```

```python
f = open('dataLR.txt', 'r')
N=10000
C=10
x = [[0 for j in range(C)] for i in range(N)] 
y = [0 for i in range(N)] 
# Loop over lines and extract variables of interest
i=-1
for line in f:
    i=i+1
    line = line.strip()
    columns = line.split(',')
    y[i]=float(columns[0])
    for j in range(0,10):
        x[i][j] = float(columns[j+1])

f.close()
```

```python
def lis_1(lis_num):    
    x_lis_1 = [-1*i for i in x[lis_num]]
    x_lis_1.insert(0,1)
    x_lis_1.append(-1)
    return x_lis_1

def lis_2(lis_num):    
    x_lis_2 = [i for i in x[lis_num]]
    x_lis_2.insert(0,1)
    x_lis_2.append(1)
    return x_lis_2
constraints_min = [lis_1(lis) for lis in range(len(x))]
constraints_max = [lis_2(lis) for lis in range(len(x))]
y10000 = [-1*i for i in y]
A = []
for i in constraints_min:
    A.append(i)
for i in constraints_max:
    A.append(i)
v = np.concatenate((np.identity(10000), np.identity(10000)), axis = 0)
new_A = np.concatenate((A,v), axis = 1)
new_A = new_A[:,1:]
```

```python
c = y10000 + y
obj_sense = GRB.MAXIMIZE
sense = [GRB.EQUAL]*10011
rhs = np.concatenate((np.zeros(11), np.ones(10000)), axis = 0)
lb = [0]*20000
ub = [GRB.INFINITY]*20000
vtype = [GRB.CONTINUOUS]*20000
sol = [0]*20000
A = np.transpose(new_A)
```

```python
success = lp_optimize(10011, 20000, c, obj_sense, A, sense, rhs, lb, ub, vtype, sol)
```
